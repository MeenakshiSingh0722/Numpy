Iteration

```python
import numpy as np
a1=np.array([1,2,3,4])
print(a1)
for i1 in a1
    print(i1)
```

```python
a2=np.array([[1,2,3],[4,5,6]])
print(a2)
for i2 in a2:
    print(i2)
print()
for i3 in a2:
    for i4 in i3:
        print(i4)
```

```python
a3=np.array([[[1,2,3],[4,5,6]]])
print(a3)
print(a3.ndim)
for i5 in a3:
    for i6 in i5:
        for i7 in i6:
            print(i7)
        
```

nditer fn for interation 

```python
a4=np.array([[[1,2,3],[4,5,6]]])
for i8 in np.nditer(a4,flags =['buffered'],op_dtypes=["S"]):
    print(i8)
```

```python
a5=np.array([[[1,2,3],[4,5,6]]])
for i9,d in np.ndenumerate(a4,):
    print(i9,d)
```

Joining

Stack fn

```python
c1=np.array([1,2,3,4])
c2=np.array([5,6,7,8])

c3=np.stack((c1,c2),axis=0)
c4=np.hstack((c1,c2))
c5=np.vstack((c1,c2))
c6=np.dstack((c1,c2))
print(c3)
print(c4)
print(c5)
print(c6)
```

Inverse()

```python
y9=np.matrix([[1,2],[4,5]])
print(np.linalg.inv(y9))
```

Power -> np.linalg.matrix_power(var,n)
           n=0 -> I
           n>0 ->power (multi)
           n<0 ->inverse*power

```python
y10=np.matrix([[1,2],[4,5]])
print(np.linalg.matrix_power(y10,2))
print(np.linalg.matrix_power(y10,0))
print(np.linalg.matrix_power(y10,-2))
```
