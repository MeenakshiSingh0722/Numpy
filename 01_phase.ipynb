Array Properties

```python
import numpy as np
ar=np.array([[1,2],[3,4]])
print("shape: ",ar.shape)
print("dimension: ",ar.ndim)
print("size: ",ar.size)
print("Data type: ",ar.dtype)
```

array reshaping

```python
arr=np.arange(12)
print("Original array: ",arr)

reshaped =arr.reshape((3,4))
print("reshaped array:", reshaped)

flattend=reshaped.flatten()
print("\n flattend array is:", flattend)

# ravel(returns view instead of copy )
raveled_array=reshaped.ravel()
print("\n reveled array: ",raveled_array)

# transpose
transposed=reshaped.T
print("\n transpose array: ",transposed)
```
