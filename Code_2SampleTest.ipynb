```python
import numpy as np
from scipy import stats
from Code_Ttest import t_stats, p_value
```

```python
grp_A=[85,88,90,92,87,85,89,91,86,88]
grp_B=[82,84,80,83,81,79,78,85,84,83]
```

```python
t_stats,p_value=stats.ttest_ind(grp_A,grp_B,equal_var=False)
t_stats
```

```python
alpha=0.05
```

```python
if p_value<alpha:
    print('reject the null hypothesis')
else:
    print('accept the null hypothesis')
```
