
Arithmetic operation

```python
import numpy as np
a1=np.array([1,2,3])
a=a1+2
print(a)
```

```python
a2=np.array([1,2,3])
a3=np.array([1,2,3])
varr=a2+a3
varr2=np.add(a2,a3)
print(varr)
```

```python
A=np.array([1,2,3])
B=np.array([1,2,3])
A1=np.add(A,B)
A2=np.subtract(A,B)
A3=np.multiply(A,B)
A4=np.divide(A,B)
A5=np.mod(A,B)
A6=np.reciprocal(A)
print(A1,A2,A3,A4,A5,A6)
```

2D Array

```python
a4=np.array([[1,2,3],[4,5,6]])
a5=np.array([[1,2,3],[4,5,6]])
varadd=a4+a5
print(varadd)
```

MACHINE LEARNING

Arithmetic fn

```python
a6=np.array([1,2,3,4,3,2,1,4,5,3])
print("max: ",np.max(a6))
print("min: ",np.min(a6))
print("position of max argument: ",np.argmax(a6))
print("position of min argument: ",np.argmin(a6))
print("sqr root: ",np.sqrt(a6))
print("sin: ",np.sin(a6))
print("cos: ",np.cos(a6))
print("community sum: ",np.cumsum(a6))
```

```python
aa=np.array([[1,2,3],[5,4,6]])
print(np.min(aa,axis=0))
print(np.max(aa,axis=1))
```

Reshape

```python
b2=np.array([1,2,3,4,5,6])
b3=b2.reshape(3,2)
print(b3)
print(b3.ndim)
```

```python
b4=np.array([1,2,3,4,5,6,7,9,11,4,7,5])
print(b4)
print(b4.ndim)
print()

b5=b4.reshape(2,3,2)
print(b5)
print(b5.ndim)
```

```python
b4=np.array([1,2,3,4,5,6,7,9,11,4,7,5])
print(b4)
print(b4.ndim)
print()

b5=b4.reshape(2,3,2)
print(b5)
print(b5.ndim)
print()

one=b5.reshape(-1)
print(one)
print(one.ndim)
```

Broadcasting

```python
c1=np.array([1,2,3])
print(c1.shape)
print(c1)
c2=np.array([[1],[2],[3]])
print(c2.shape)
print(c2)
print(c1+c2)
```

```python
c3=np.array([[1],[2]])
print(c3.shape)
c4=np.array([[1,2,3],[1,2,3]])
print(c4.shape)
print(c3+c4)
```

Indexing

Slicing

```python
c7=np.array([9,8,7,6,4,5,6,7,9])
print(c7)
print(c7[1:3])
print(c7[1:])
print(c7[:4])
print("stop:",c7[::2])
```
