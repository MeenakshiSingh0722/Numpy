data type in numpy array

```python
a1=np.array([1,2,3,4])
print("Data type:",a1.dtype)
```

change datatype

```python
a2=np.array([1,2,3,4],dtype=np.int8)
print("Data type:",a2.dtype)
```

with shortcut

```python
a3=np.array([1,2,3,4],dtype="f")
print("Data type:",a3.dtype)
```

with function

```python
a3=np.array([1,2,3,4])
new=np.float32(a3)
new_one=np.int_(new)

print("Data type:",a3.dtype)
print("Data type:",new.dtype)
print("Data type:",new_one.dtype)

print(a3)
print(new)
print(new_one)
```

directly convert

```python
a4= np.array([1,2,3,4])
new1=a4.astype(float)
print(a4)
print(new1)
```
