```python
import numpy as np
from scipy.stats import norm
```

```python
sample=[172,174,168,169,171,173,175,170,169,172]
# we know sample data must be greater than 30 it's just for example
popln_mean=170
popln_std=3
samp_mean=np.mean(sample)
n=len(sample)
```

```python
z_score=(samp_mean-popln_mean)/(popln_std/np.sqrt(n))
z_score
```

```python
p_value=2*(1-norm.cdf(abs(z_score)))
p_value
```

```python
alpha=0.05
```

```python
 if p_value<alpha:
    print("i'll reject the null hypothesis")
else:
    print("i'll accept the null hypothesis")
```
