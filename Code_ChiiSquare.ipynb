```python
import numpy as np
import seaborn as sns
from scipy.stats import chi2_contingency
import pandas as pd
```

```python
df=sns.load_dataset('titanic')
df
```

```python
contigency_table=pd.crosstab(df['sex'],df['survived'])
contigency_table
```

```python
chi2,p_value,dof,expected=chi2_contingency(contigency_table)
expected
```

```python
p_value
```

```python
chi2
```

```python
alpha=0.05
```

```python
if p_value<alpha:
    print('reject the null hypothesis and there is a significant relationship between gender and survive')
else:
    print('accept the null hypothesis and there is a no significant relationship between gender and survive')
```
